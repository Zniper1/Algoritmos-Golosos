```python

# Calculamos el precio por metro cuadrado para cada barrio
inmuebles_por_barrio = inmuebles.groupby('Barrio').mean()
inmuebles_por_barrio['Precio_m2'] = inmuebles_por_barrio['Precio_Millon'] / inmuebles_por_barrio['Area']
inmuebles_por_barrio.sort_values(by='Precio_m2', ascending=False, inplace=True)

# Creamos un gráfico de barras con Seaborn
plt.figure(figsize=(8,27))
grafica = sns.barplot(x='Precio_m2', y=inmuebles_por_barrio.index, data=inmuebles_por_barrio)

# Agregamos las etiquetas del precio por metro cuadrado en cada barra
for i, precio in enumerate(inmuebles_por_barrio['Precio_m2']):
    grafica.text(precio+0.5, i, '${:.2f}'.format(precio), va='center')

# Configuramos los títulos y etiquetas del gráfico
grafica.set_title('Precio por metro cuadrado en cada barrio')
grafica.set_xlabel('Precio por metro cuadrado (millones de pesos)')
grafica.set_ylabel('Barrio')

plt.tight_layout()
plt.show()
```

```python
#PRECIO M2 POR BARRIO
#Se crea una columna "Valorm2" Se toma el valor total y se dibide por el area
inmuebles["Valor_m2_Millon"] = inmuebles["Precio_Millon"]/inmuebles["Area"]

#Calculo el m2 para cada uno de los barrio//
#groupby es agrupar//Mean es el promedio
inmuebles.groupby("Barrio").mean()
#Para todas las columnas numericas trae el resultado el promedio

```

```python

#Ahora vamos a sumar todos los valores de los precios de un barrio y lo vamos a 
#dividir por todas las superficies ya que daria el resultado exacto

datos_barrio = inmuebles.groupby("Barrio").sum()
datos_barrio
```

```python
#Ahora calculamos el valor exacto, ya se encuentra sumado todos las areas
#y anteriormente se saco el promedio
datos_barrio["Valor_m2_Barrio"] = datos_barrio["Precio_Millon"]/datos_barrio["Area"]
datos_barrio
#Respecto el resultado anterior Valor_m2_Millon
```

```python
#Tomamos el valor correcto Valor_m2_Barrio y lo sustituimos por el valor
#promedio de los metros cuadrados "Valor_m2_Millon"
#Se generara un diccionario para remplazar las columnas
#****dict(datos_barrio["Valor_m2_Barrio"])***
#imprime los barrios y el valor del m2 por barrio.(El correcto)
#lo ponemos dentro de una variable
m2_barrio = dict(datos_barrio["Valor_m2_Barrio"])
#Tomo la columna Valor_m2_Millon y le asigno los barrios
inmuebles["Valor_m2_Barrio"] = inmuebles["Barrio"]
#Le voy a hacer un maping , tomara todos los valores del diccionario y lo pondra
#dentro de la variable(Diccionario)
inmuebles["Valor_m2_Barrio"] = inmuebles["Valor_m2_Barrio"].map(m2_barrio)

inmuebles.head(5)
```

```python
#Consultas para analizar, buscando los primeros 10 barrios con mas inmuebles 
#a la venta y los almaceno en una variable  
top_barrios = inmuebles["Barrio"].value_counts()[:10].index
#se realiza un reset para que no quede "Barrio" como indice
#Inplace nos permite realizar la ejecucion sin la necesidad de aplicarlo a otra
#Variable
datos_barrio.reset_index(inplace=True)
```

```python
#Consulto los barrios que esten en top barrio// la @ (Si se encuentra)
datos_barrio.query("Barrio in @top_barrios")
#Me devuelve el top 10 de los barrios en un Data Frame
```

```python
#Ahora es el momento del maching Learning
#importamos nuestro paquete (modelo linear ), 
from sklearn.linear_model import LinearRegression
#importamos paquete de separacion de entrenamiento  y pruebas
from sklearn.model_selection import train_test_split
#introducimos en el X una matriz y verificamos cuanto saldriia el inmueble en esa zona
#Ingresamos como variables el precio en millones de los inmuebles
#y la zona upz. 
X = datos_ml[['COD_UPZ_GRUPO']]
y = datos_ml['Precio_Millon']

# Necesito 4 variables, 2 para entrenamiento y 2 de prueba // split de ambos data set
#el test size por defecto es el 0.25, y tengo que realizar un estado de alotoriedad de mis datos
#que entrene 75% de los datos y pruebe con 25% d los datos
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=99)
```

```python
#instanciamos una formacion linear de forma generica  
modelo = LinearRegression()
```

```python
#agustamos la x de entrenamiento y la y de entrenamiento
modelo.fit(X_train,y_train)
```

```python
#se almacena el resultado en y predict.
y_predict_test = modelo.predict(X_test)
```

```python
#necesitamos ver el error absoluto medio, es tomar la diferencia de valores que podemos encontrar
#para hacer el calculo, 
#importamos 
from sklearn.metrics import mean_absolute_error, r2_score

baseline_mae = mean_absolute_error(y_test, y_predict_test)
baseline_r2 = r2_score(y_test, y_predict_test)

print(baseline_mae,baseline_r2)
```
